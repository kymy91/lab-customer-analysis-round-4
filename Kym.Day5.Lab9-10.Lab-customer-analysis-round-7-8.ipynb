```python
#removing outliers
iqr = np.percentile(numbers_final['Total Claim Amount'],75) - np.percentile(numbers_final['Total Claim Amount'],25)
upper_limit = np.percentile(numbers_final['Total Claim Amount'],75) + 1.5*iqr
lower_limit = np.percentile(numbers_final['Total Claim Amount'],25) - 1.5*iqr
```

```python
data = data[(numbers_final['Total Claim Amount']>lower_limit) & (numbers_final['Total Claim Amount']<upper_limit)]
```
